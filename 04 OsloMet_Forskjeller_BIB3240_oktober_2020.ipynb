```python
ddk_krig = nb.book_urn(ddk='355%', period = (1990, 2010), limit=50)
ddk_krig
```

```python
tot = nb.frame_sort(nb.frame(nb.totals ( 50000), 'tot'))
```

```python
tot.head(20)
```

```python
ddk_krig = nb.frame(nb.aggregate_urns(ddk_krig), 'ddk_krig')
```

```python
nb.frame_sort(ddk_krig).head(20)
```

```python
nb.normalize_corpus_dataframe(tot)
```

```python
nb.normalize_corpus_dataframe(ddk_krig)
```

```python
nb.frame_sort(nb.frame(ddk_krig['ddk_krig']/tot['tot'])).head(50)
```

```python
subject_krig = nb.book_urn(subject = 'krig', period = (1990, 2010), limit = 50 )
subject_krig
```

```python
subject_krig = nb.frame(nb.aggregate_urns(subject_krig), 'subject_krig')
```

```python
nb.normalize_corpus_dataframe(subject_krig)
```

```python
nb.frame_sort(nb.frame(subject_krig['subject_krig']/tot['tot'])).head(50)
```

```python
innhold_krig = nb.book_urn(words = ['krig', 'Krig'], period = (1990, 2010), limit = 50)
innhold_krig
```

```python
innhold_krig = nb.frame(nb.aggregate_urns(innhold_krig), 'krig')
nb.normalize_corpus_dataframe(innhold_krig)
nb.frame_sort(nb.frame(innhold_krig['krig']/tot['tot'])).head(50)
```

```python
title_krig = nb.book_urn(title = '%krig%', period = (1990, 2010), limit = 50 )
title_krig = nb.frame(nb.aggregate_urns(title_krig), 'krig')
nb.normalize_corpus_dataframe(title_krig)
nb.frame_sort(nb.frame(title_krig['krig']/tot['tot'])).head(50)
```
